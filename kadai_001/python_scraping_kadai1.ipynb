```python
!pip install requests beautifulsoup4
import requests
from bs4 import BeautifulSoup
url = 'https://news.yahoo.co.jp/expert/articles/37093578100376494b93c5a0bcb1a9c39aa4e11a'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
main_element = soup.select('#uamods-article > div:nth-child(1) > section > div > p')
main_text = ''.join([i.text for i in main_element])
print(main_text)
```
