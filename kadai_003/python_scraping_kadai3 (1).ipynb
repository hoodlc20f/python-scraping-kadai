```python
!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from getpass import getpass

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# chromeを開く
chrome_driver = webdriver.Chrome(options=chrome_options)

# 最大30秒待つ
wait = WebDriverWait(chrome_driver, 30)

# 侍テラコヤトップページ取得
chrome_driver.get('https://terakoya.sejuku.net/')

# ログインボタン取得→クリック
header_login_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR, '.sc-dUnjic.sc-ddHOmm.gctkGe.gouqxP')
    )
)
header_login_button.click()

# メアド、パスワード入力
email_address = input('メールアドレスを入力してください： ')
password = getpass('パスワードを入力してください： ')
parent_element = chrome_driver.find_element(By.CSS_SELECTOR, '.sc-dfjikj.gjkOYo')
email_input = parent_element.find_element(By.NAME, 'email')
password_input = parent_element.find_element(By.NAME, 'password')
email_input.send_keys(email_address)
password_input.send_keys(password)

# フォームログインボタン取得→クリック
form_login_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR, '#root > div.sc-jHbxoU.cBQDng > div.sc-jZCRgm.jqrmih > div.sc-cMHOsR.ehuTZK > button')
    )
)
form_login_button.click()

# アイコンボタン取得→クリック
icon_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR, '#root > div.sc-kATJUg.ldpnLv > div > div > header > div.sc-jBBmjq.kIGIev > div.sc-gHZRzJ.hcYyZT > div')
    )
)
icon_button.click()

# アカウント設定ボタン取得→クリック
account_setting_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR, '#root > div.sc-kATJUg.ldpnLv > div > div > header > div.sc-jhzBug.WpVTw > ul > a:nth-child(1) > li')
    )
)
account_setting_button.click()

# 編集ボタン取得→クリック
edit_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR, '.sc-dDvxFM.bTLjQy.sc-jBTeXN.fFGUyp')
    )
)
edit_button.click()

# 自己紹介文の自動入力
self_intro = 'プログラミング学習中です！今はスクレイピングに挑戦しています！'
self_input = chrome_driver.find_element(By.CSS_SELECTOR, 'div:nth-child(9) > div.sc-dETfiT.deCxzg > textarea')
self_input.clear()
self_input.send_keys(self_intro)

# 更新ボタンの取得→クリック
update_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR, '.sc-dDvxFM.dAeCag.sc-cdSzGR.cmfwEQ')
    )
)
update_button.click()

# ブラウザ終了
chrome_driver.quit()
```
